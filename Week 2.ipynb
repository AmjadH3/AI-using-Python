### Instance Variable and Methods
Instance - Object created from Class  
. Every object has a own attribute and methods and use them for themselves

```python
class Student:
    def __init__(self, name, age, grade):  #All the methods which have self object passed are the Intance Methods.
        self.name=name     #These are 3 instance variable initializing 
        self.age=age       #these start with 'self.' and initize with variable passed in constructor(init) fucn e.g name, age.
        self.grade=grade
    def displayInfo(self):  #this is also self object passed
        print(f"Student Name: {self.name}")  # these are instance var that is why we need to use self. Var
        print(f"Student age: {self.age}")
        print(f"Student Grade:{self.grade}")
    def is_eligible (self):  # If we dont pass self then that method or attribute dont belong to that particular Object.
        if self.age >= 15:
            print (self.name, "is eligible for admission")
        else:
            print(self.name, " is not eligible for admission")
```

```python
me=Student('Amjad', 30, '16th')
me.displayInfo()
me.is_eligible()
```

### Encapsulation
keeping all the data and methods inside that object, can manupilate the attribute using methods,  
Can hide the data from user. and are kept in the class and the methods of class is exposed.  
Private data- can be only used in particular class using the methods not in global scale

```python
class Student:
    def __init__(self, name, grade):
        self.__grade = grade   # private attribute
student1 = Student('Ali', 90)
print(student1.__grade)       # Error
```

So in the above case get help from class methods to access the hidden data, which is following

```python
class Student:
    def __init__(self, name, grade):
        self.__grade = grade

    def get_grade(self):     # a seperate func to access the private data to outside world
        return self.__grade
        
student1 = Student('Ali', 90)
print(student1.get_grade())       
```

```python
class Student:
    @property
    def grade(self):
        return self.__grade
        
    def __init__(self, name, grade):
        self.__grade = grade

    def get_grade(self):    
        return self.__grade

    def set_grade(self, grade):   #setter is used for the logical input, in this case marks should be btw 0-100.
        if 0 <= grade <=100:
            self.__grade = grade
        else:
            print('Invalid grade!')
            
        
student1 = Student('Ali', 90)
student1.set_grade(-95)
student1.get_grade()
```

```python
student1.set_grade(95)
student1.get_grade()
```

### Topic 39- Python Special Methods
2 concept interchangably used 1 func(globally) 2 method (inside class)  
3rd is Special func - builtin behavior auto associated with certain operations  
  behavior of func which is already present in common func e.g can print any var,
when create class so many elements and py don't know how to print so py give us facility to tell class to print in which sequence we want  
e.g print, length, so py don't know how to print diff element and we can tell how to do it  
To make a function special use underscore before and after with the Method  
- Special Method is default except it is modified according to need

```python
class Book:
    def __init__(self, title, author):
        self.title=title
        self.author=author
    def __str__(self):  # undercover shows that we are redefining a default function of str.
        return f"'{self.title}' by {self.author}"   # str func will treat every object as a string
```

```python
b= Book('1987','George')
print(b)
c = Book(1987,'George')
print(c)
```

The above methods are default func or methods, we are telling python how our book will be used as a string.  
 The str func has return value in double quotes and will must be string.  
Before this we discussed user defined func.

```python
class Playlist:
    def __init__(self, songs):
        self.songs=songs
    def __len__(self):
        return len(self.songs)
    def __add__(self, other):
        return Playlist(self.songs + other.songs)
```

```python
p1 = Playlist(["Song1", "Song2"])
p2 = Playlist(["Song3"])
print(len(p1))
print(len(p1 + p2))
```

```python
class Point:
    def __init__ (self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return f"({self.x} , {self.y})"
    def __add__(self, other):
        if not isinstance (other, Point):
            return NotImplemented
        return Point(self.x + other.x, self.y + other.y)
```

```python
p1 = Point(2, 3)
p2 = Point(4, 1)
print(p1+p2)
```

### Topic 47 Creating Custom Exceptions in PYthon
- If there is no exveptions in python by default we want to raise, so we should make a custom one and give a descriptive name
- You have to inherit custom exception from exception class to get all functionalities of exception class.

```python
class InvalidAgeError(Exception):  # We inherit this exception from exception class
    pass
```

```python
age = int(input("Enter age: "))
if age < 0:
    raise InvalidAgeError(f"Age cannot be negative!, {age}")
```

So now we need to handle this error, and it is same as default error handling.

```python
try:
    age=int(input("Enter age: "))
    if age < 0:
        raise InvalidAgeError(f"Age can't be negative! , {age}")
    
except InvalidAgeError as e:
    print(e)
```
