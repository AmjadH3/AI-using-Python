```python
i = 4
row= i
col= i
matrix = []

for i in range(row):
    row = []
    for c in range(col):
        row.append(0)
    matrix.append(row)

for row in matrix:
    print (' '.join(map(str, row)))
```

```python
matrix = [[0 for _ in range(4)] for _ in range(4)]   
for row in matrix:
    print (' '.join(map(str, row))) #same code as above 
#row = [0, 0, 0, 0]
      #   ↓
    #map(str, row)
     #    ↓
    #['0', '0', '0', '0']   ← all strings now
   #      ↓
#' '.join(...)
  #       ↓
 #  '0 0 0 0'              ← single string with spaces //
    # JOin only works with strings thats why first need to convert to string
```

```python
 # Dictionaries - Looping
a = {
    'a' : 20,
    'b' : 'z',
    'c' : 7
}
for key in a:
    print(key)
```

```python
for key in a:
    print(a[key])
```

```python
for key in a:
    print (key,':', a[key])
```

```python
for key,values in a.items():
    print(key,':',values)
```

Lambda- Fuction

```python
add = lambda a,b: a+b
print (add(4,6))
```

```python
doubleIt = lambda n: n*2
print (doubleIt(3))
```

```python
nums = [1,2,3,4]
square=list(map(lambda x: x*x, nums))
```

```python
print (square)
```

```python
numbs = [1,2,3,4,5,6,7]
oddNum = list(filter(lambda x: x % 2 != 0, numbs)) #filter builtin function
```

```python
print(oddNum)
```
